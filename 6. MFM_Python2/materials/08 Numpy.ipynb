## Numpy For image Processing 

```python
import cv2
import matplotlib.pyplot as plt
```

### Read Image

```python
# read image
image = cv2.imread("love.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print (image.shape)
plt.imshow(image)
plt.show()
```

###  Image Grayscaling

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image2, cmap='gray')
plt.show()
```

### Numpy Slicing

```python
sliced_image = image[0:250, 0:250, :]
plt.imshow(sliced_image)
plt.show()
```

```python
B = image[:, :, 2]
plt.imshow(B, 'gray')
plt.show()
```

### Numpy Vstack and Hstack

```python
hStack = np.hstack((sliced_image, sliced_image))
plt.imshow(hStack)
plt.show()
```

```python
vStack = np.vstack((sliced_image, sliced_image))
plt.imshow(vStack)
plt.show()
```

### Image Features Extraction
Several algorithms for extracting image features:
1. PCA
2. LBP
3. HoG
4. Transfer learning with Inception-v3, resnet, VGG, etc

#### Example: Extracting Image Features with PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
pca_features = pca.fit_transform(image[:, :, 0])
print (pca_features)
```

## More Serious

```python
ktp_ori = cv2.imread("ktp.jpg")
ktp = cv2.cvtColor(ktp_ori, cv2.COLOR_BGR2RGB)
ktp = cv2.cvtColor(ktp, cv2.COLOR_BGR2GRAY)
print (ktp.shape)
plt.imshow(ktp, cmap='gray')
plt.show()
```

### Now, we want to segment the text

#### Try with: ADAPTIVE_THRESH_GAUSSIAN threshold

![gaus](gaus.jpg)

```python
th3 = cv2.adaptiveThreshold(ktp, 191, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
plt.imshow(th3, cmap='gray')
plt.show()
```

#### Result: BAD
#### Try with: OTSU threshold

![otsu](otsu.jpg)

```python
ret3,th4 = cv2.threshold(ktp,0,  255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(th4, cmap='gray')
plt.show()
```

#### Try Dilation
![title](dilation.jpg)

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(th4, kernel, iterations = 4)

plt.imshow(dilation, cmap='gray')
plt.show()
```
